```python
df['selling_price'] = df['selling_price'].fillna(df['selling_price'].median())
```

```python
df['delivery date'] = df['delivery date'].fillna(df['delivery date'].median())
df['item_date'] = df['item_date'].fillna(df['item_date'].median())
df['status'] = df['status'].fillna(df['status'].mode()[0])
```

```python
df[df['id'].isna()]
```

```python
mode_id = df['id'].mode()[0]
df.loc[df['id'].isna(),'id'] = df.loc[df['id'].isna()].index.astype(str) + '_' + mode_id
```

```python
df.isna().sum()
```

```python
df['item_date'] = pd.to_datetime(df['item_date'], format='%Y%m%d', errors='coerce')
```

```python
df['delivery date'] = pd.to_datetime(df['delivery date'],format='%Y%m%d',errors='coerce')
```

```python
df['delivery date'] = df['delivery date'].fillna(df['item_date'])
```

```python
df['item_date'] = df['item_date'].fillna(df['delivery date'])
```

```python
df['quantity tons'] = pd.to_numeric(df['quantity tons'], errors='coerce')
```

```python
import numpy as np

df2_skew = df2_filtered.copy()
# Apply log transformation to skewed features
for feature in skewed_features_before.index:
    df2_skew[feature] = np.log1p(df2_filtered[feature])

# Identify skewed continuous features again after transformation
skewed_features_after = df2_skew.select_dtypes(include=['float64']).apply(lambda x: x.skew()).sort_values(ascending=False)
skewed_features_after = skewed_features_after.drop(index='customer')

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle

X= df3.drop(['selling_price','status'],axis=1)
Y_regression = df3['selling_price']
Y_classification =df3['status']


#standard Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

with open('scaler.pkl','wb') as f:
    pickle.dump(scaler, f)
```

```python
with open('X.pkl','wb') as f:
    pickle.dump(X, f)
```

```python
X_train_reg, X_test_reg, Y_train_reg, Y_test_reg = train_test_split(X_scaled, Y_regression, test_size=0.2, random_state=42)
X_train_clf, X_test_clf, Y_train_clf, Y_test_clf = train_test_split(X_scaled, Y_classification, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import ExtraTreesRegressor, ExtraTreesClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, precision_score, recall_score,f1_score, roc_curve, roc_auc_score

Extree_reg_model = ExtraTreesRegressor()
Extree_reg_model
```

```python
Extree_reg_model.fit(X_train_reg,Y_train_reg)
```

```python
with open('Extree_reg_model.pkl','wb') as f:
    pickle.dump(Extree_reg_model, f)
```

```python
Y_pred_reg = Extree_reg_model.predict(X_test_reg)
```

```python
# R-squared value
print('r-squared :',r2_score(Y_test_reg, Y_pred_reg))
print('mean_squared :',mean_squared_error(Y_test_reg, Y_pred_reg))
print('mean_avsolute :',mean_absolute_error(Y_test_reg, Y_pred_reg))
```
